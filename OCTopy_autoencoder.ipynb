```python
class Encoder(nn.Module):
    def __init__(self):
        super(Encoder, self).__init__()
        self.encoder = nn.Sequential( # 224 x 224
            nn.Conv2d(3, 16, kernel_size=3, padding = 1), # 222 x 222
            nn.LeakyReLU(0.4),
            nn.MaxPool2d(2, 2), # 111 x 111
            nn.BatchNorm2d(16),
            nn.Conv2d(16, 32, kernel_size=3, padding=1), # 110 x 110
            nn.LeakyReLU(0.4),
            nn.Dropout(0.6),
            nn.MaxPool2d(2, 2), # 55 x 55
            nn.BatchNorm2d(32),
            nn.Conv2d(32, 64, kernel_size=3, padding=1), # 54 x 54
            nn.Dropout(0.5),
            nn.LeakyReLU(0.4),
            nn.MaxPool2d(2, 2), # 27 x 27
            nn.BatchNorm2d(64),
            nn.Conv2d(64, 128, kernel_size=3, padding=1), # 26 x 26
            nn.LeakyReLU(0.4),
            nn.MaxPool2d(2, 2), # 13 x 13
            nn.BatchNorm2d(128),
            nn.Conv2d(128, 256, kernel_size=3, padding=1), # 12 x 12
            nn.LeakyReLU(0.4),
            nn.Dropout(0.4),
            nn.BatchNorm2d(256),
            nn.Conv2d(256, 512, kernel_size=3, padding=1), # 11 x 11
            nn.LeakyReLU(0.4),
            nn.Dropout(0.3),
            nn.BatchNorm2d(512),
            nn.Conv2d(512, 1024, kernel_size=3, padding=1), # 10 x 10
            nn.LeakyReLU(0.4),
            nn.MaxPool2d(2, 2), # 5 x 5
            nn.BatchNorm2d(1024),
            nn.Conv2d(1024, 2048, kernel_size=3, padding=1), # 4 x 4
            nn.LeakyReLU(0.4)
        )

    def forward(self, x):
        x = self.encoder(x)
        return x

# Decoder class
class Decoder(nn.Module):
    def __init__(self):
        super(Decoder, self).__init__()
        self.decoder = nn.Sequential( # 4 x 4
            nn.ConvTranspose2d(2048, 1024, kernel_size=3, padding=1), # 5 x 5
            nn.LeakyReLU(0.4),
            nn.Dropout(0.3),
            nn.Upsample(scale_factor=2, mode='nearest'), # 10 x 10
            nn.BatchNorm2d(1024),
            nn.ConvTranspose2d(1024, 512, kernel_size=3, padding=1), # 11 x 11
            nn.LeakyReLU(0.4),
            nn.Dropout(0.3),
            nn.BatchNorm2d(512),
            nn.ConvTranspose2d(512, 256, kernel_size=3, padding=1), # 12 x 12
            nn.LeakyReLU(0.4),
            nn.BatchNorm2d(256),
            nn.ConvTranspose2d(256, 128, kernel_size=3, padding=1), # 13 x 13
            nn.LeakyReLU(0.4),
            nn.Upsample(scale_factor=2, mode='nearest'), # 26 x 26
            nn.BatchNorm2d(128),
            nn.ConvTranspose2d(128, 64, kernel_size=3, padding=1), # 27 x 27
            nn.LeakyReLU(0.4),
            nn.Upsample(scale_factor=2, mode='nearest'), # 54 x 54
            nn.Dropout(0.3),
            nn.BatchNorm2d(64),
            nn.ConvTranspose2d(64, 32, kernel_size=3, padding=1), # 55 x 55
            nn.LeakyReLU(0.4),
            nn.Upsample(scale_factor=2, mode='nearest'), # 110 x 110
            nn.Dropout(0.3),
            nn.BatchNorm2d(32),
            nn.ConvTranspose2d(32, 16, kernel_size=3, padding=1), # 111 x 111
            nn.LeakyReLU(0.4),
            nn.Upsample(scale_factor=2, mode='nearest'), # 222 x 222
            nn.BatchNorm2d(16),
            nn.ConvTranspose2d(16, 3, kernel_size=3, padding=1), # 224 x 224
            nn.LeakyReLU(0.3)
        )

    def forward(self, x):
        x = self.decoder(x)
        return x

# Autoencoder class combining Encoder and Decoder
class Autoencoder(nn.Module):
    def __init__(self, encoder, decoder):
        super(Autoencoder, self).__init__()
        self.encoder = encoder
        self.decoder = decoder

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

# Create instances of Encoder and Decoder
encoder = Encoder()
decoder = Decoder()

# Create an instance of the Autoencoder
autoencoder = Autoencoder(encoder, decoder)
```

```python
# Train linear classifier

train_loss, train_acc, test_loss, test_acc, val_loss, val_acc = [], [], [], [], [], []

classifier_model = classifier_model.to(device)
# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(classifier_model.parameters(), lr=0.001)
# base_optim = RAdam(autoencoder.parameters(), lr = 0.001)
# optimizer = Lookahead(base_optim, k=5, alpha=0.5)

# Assuming PyTorch tensors are the input data
# If not, need to convert input data into PyTorch tensors
# For example, using torchvision.transforms.ToTensor()

# Assuming you have train_data and test_data PyTorch tensors

# Training the Autoencoder
num_epochs = 200
for epoch in range(num_epochs):
    total_training_loss = 0.0
    total_training_accu = 0.0
    classifier_model.train()

    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)
        # Forward pass
        outputs = classifier_model(images)
        # print("output dim", outputs.shape)
        # print("labels dim", labels.shape)
        loss = criterion(outputs, labels)
        preds = torch.argmax(outputs, dim=1)
        accu = torch.sum(preds == labels) / len(labels)
        total_training_accu += accu

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_training_loss += loss.item()

    average_training_loss = total_training_loss / len(train_loader)
    average_training_accu = (total_training_accu / len(train_loader)) * 100
    print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {average_training_loss}, Train Accu: {average_training_accu}', end='  |  ')
    train_loss.append(average_training_loss)
    train_acc.append(average_training_accu)

    # Testing
    classifier_model.eval()
    total_test_loss = 0.0
    total_test_accu = 0.0
    with torch.no_grad():
        for images, labels in test_loader:
            # Forward pass
            images = images.to(device)
            labels = labels.to(device)
            outputs = classifier_model(images)
            loss = criterion(outputs, labels)
            total_test_loss += loss.item()
            preds = torch.argmax(outputs, dim=1)
            accu = torch.sum(preds == labels) / len(labels)
            total_test_accu += accu

    average_test_loss = total_test_loss / len(test_loader)
    average_test_accu = (total_test_accu / len(test_loader)) * 100
    test_loss.append(average_test_loss)
    test_acc.append(average_test_accu)

    print(f'Test Loss: {average_test_loss}, Test Accu: {average_test_accu}', end='  |  ')

    # Validation
    classifier_model.eval()
    total_validation_loss = 0.0
    total_validation_accu = 0.0
    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)
            # Forward pass
            outputs = classifier_model(images)
            loss = criterion(outputs, labels)
            total_validation_loss += loss.item()
            preds = torch.argmax(outputs, dim=1)
            accu = torch.sum(preds == labels) / len(labels)
            total_validation_accu += accu

    average_validation_loss = total_validation_loss / len(val_loader)
    val_loss.append(average_validation_loss)

    average_val_accu = (total_validation_accu / len(val_loader)) * 100
    val_acc.append(average_val_accu)

    print(f'Val Loss: {average_validation_loss}, Val Accu: {average_val_accu}')

    if epoch > 0 and epoch % 20 == 0:
        torch.save(classifier_model.state_dict(), f'{ROOT_PATH}/classifier_epoch_{epoch}.ckpt')
        # torch.save(classifier_model.encoder.state_dict(), f'{ROOT_PATH}/linear_epoch_{epoch}.ckpt')

        fname = f"{ROOT_PATH}/linear_loss_epoch_{epoch}.pkl"
        output = {"train_loss" : train_loss, "test_loss" : test_loss, "val_loss" : val_loss}
        with open(fname, 'wb') as outfile:
            pk.dump(output, outfile)

        fname = f"{ROOT_PATH}/linear_accu_epoch_{epoch}.pkl"
        output = {"train_accu" : train_acc, "test_accu" : test_acc, "val_accu" : val_acc}
        with open(fname, 'wb') as outfile:
            pk.dump(output, outfile)
```

```python
class Encoder(nn.Module):
    def __init__(self):
        super(Encoder, self).__init__()
        self.encoder = nn.Sequential( # 224 x 224
            nn.Conv2d(3, 64, kernel_size=3, padding = 1), # 222 x 222
            nn.LeakyReLU(0.4),
            nn.MaxPool2d(2, 2), # 111 x 111
            nn.BatchNorm2d(64),
            nn.Conv2d(64, 128, kernel_size=3, padding=1), # 110 x 110
            nn.LeakyReLU(0.4),
            nn.Dropout(0.6),
            nn.MaxPool2d(2, 2), # 55 x 55
            nn.BatchNorm2d(128),
            nn.Conv2d(128, 192, kernel_size=3, padding=1), # 54 x 54
            nn.Dropout(0.5),
            nn.LeakyReLU(0.4),
            nn.MaxPool2d(2, 2), # 27 x 27
            nn.BatchNorm2d(192),
            nn.Conv2d(192, 256, kernel_size=3, padding=1), # 26 x 26
            nn.LeakyReLU(0.4)
        )

    def forward(self, x):
        x = self.encoder(x)
        return x

# Decoder class
class Decoder(nn.Module):
    def __init__(self):
        super(Decoder, self).__init__()
        self.decoder = nn.Sequential( # 4 x 4
            nn.ConvTranspose2d(256, 192, kernel_size=3, padding=1), # 55 x 55
            nn.LeakyReLU(0.4),
            nn.Upsample(scale_factor=2, mode='nearest'), # 110 x 110
            nn.Dropout(0.3),
            nn.BatchNorm2d(192),
            nn.ConvTranspose2d(192, 128, kernel_size=3, padding=1), # 111 x 111
            nn.LeakyReLU(0.4),
            nn.Upsample(scale_factor=2, mode='nearest'), # 222 x 222
            nn.BatchNorm2d(128),
            nn.ConvTranspose2d(128, 64, kernel_size=3, padding=1), # 111 x 111
            nn.LeakyReLU(0.3),
            nn.Upsample(scale_factor=2, mode='nearest'), # 222 x 222
            nn.BatchNorm2d(64),
            nn.ConvTranspose2d(64, 3, kernel_size=3, padding=1), # 224 x 224
            nn.LeakyReLU(0.3)
        )

    def forward(self, x):
        x = self.decoder(x)
        return x

# Autoencoder class combining Encoder and Decoder
class Autoencoder(nn.Module):
    def __init__(self, encoder, decoder):
        super(Autoencoder, self).__init__()
        self.encoder = encoder
        self.decoder = decoder

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

# Create instances of Encoder and Decoder
encoder = Encoder()
decoder = Decoder()

# Create an instance of the Autoencoder
autoencoder = Autoencoder(encoder, decoder)
```
